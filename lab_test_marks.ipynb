```python
# import csv file in pandas
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
df = pd.read_csv('./data/CP/labtest1_final.csv')
# df['Marks'] = df['Marks'].fillna(0).astype(float)
#rename total column to marks
df.rename(columns={'Total':'Marks'}, inplace=True)
# add column with middle 2 digits of id
df['code'] = df['ID No'].str[4:6]

df
```

```python
import seaborn as sns
sns.set_theme()

df.hist(column='Marks',bins=75)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution for CP Lab Test 1')
```

```python
print(df['Marks'].mean())
print(df['Marks'].median())
print(df['Marks'].mode())
```

```python
from scipy import stats

for i in range(38, 76):
    print(f"marks: {i} -> {stats.percentileofscore(df['Marks'], i, kind='weak')}")
```

```python
fig, ax = plt.subplots(figsize = (6,4))
df['Marks'].plot(kind = "hist", density=True, bins=75)
df['Marks'].plot(kind = "kde", ax=ax)
ax.set_xlim(0, 75)
plt.show()
```

```python
import matplotlib.ticker as ticker

fig, ax = plt.subplots(figsize = (10,4))
fun = lambda k: [i/k for i in list(range(1, 75*k + 1))]

a = fun(2)
b = [stats.percentileofscore(df["Marks"],i, kind='weak') for i in a]
ax.plot(a, b)

lis = [i for i in range(0, 101)]

plt.axvline(x=df['Marks'].quantile([0.5]).values[0], color='pink', linestyle='--', label='50th percentile')
plt.axvline(x=df['Marks'].quantile([0.9]).values[0], color='red', linestyle='--', label='90th percentile')
plt.axvline(x=df['Marks'].quantile([0.75]).values[0], color='green', linestyle='--', label='75th percentile')
plt.axvline(x=df['Marks'].quantile([0.8]).values[0], color='yellow', linestyle='--', label='80th percentile')



ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.set_ylim(0,101)
ax.set_xlim(0,75)
ax.set_xlabel("Marks")
ax.set_ylabel("Percentile")
ax.legend()
plt.show()
```
